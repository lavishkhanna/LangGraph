```python
from langgraph.graph import StateGraph,START,END
from typing import TypedDict
from langchain_google_genai import ChatGoogleGenerativeAI
from dotenv import load_dotenv
import os
```

```python
load_dotenv()
```

```python
llm = ChatGoogleGenerativeAI()
```

```python
class BlogState(TypedDict):
    topic:str
    outline:str
    content:str
```

```python
def gen_outline(state:BlogState)->BlogState:
    topic=state['topic']

    prompt=f"Generate a detailed outline for a blog on the topic {topic}"
    outline=llm.invoke(prompt).content
    state['outline']=outline

    return state


def gen_content(state:BlogState)->BlogState:
    outline=state['outline']

    prompt=f"generate a blog for this topic {state['topic']}, having these as outlines {state['outline']}"
    content=llm.invoke(prompt).content

    state['content']=content

    return state
```

```python
graph=StateGraph(BlogState)

graph.add_node('gen_outline',gen_outline)
graph.add_node('gen_content',gen_content)


graph.add_edge(START,'gen_outline')
graph.add_edge('gen_outline','gen_content')
graph.add_edge('gen_outline',END)

workflow=graph.compile()
```

```python
initial_state={
    "topic":"Seaship vessel contents"
}

output=workflow.invoke(initial_state)

print(output)
```
